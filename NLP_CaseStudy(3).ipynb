# **Text PreProcessing**

```python
#character removal
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt') 
nltk.download('wordnet')
```

```python
for i in range(len(df)):
  words=word_tokenize(df["News"][i].lower())
  lemmatizer=WordNetLemmatizer()
  words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
  df["News"][i]= " ".join(words)
```

# **`Model Fitting`**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection, naive_bayes, svm
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['News'] ,df['Close'].values,test_size=0.2)
```

```python
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(df['News'])

Train_X_Tfidf = Tfidf_vect.transform(x_train)
Test_X_Tfidf = Tfidf_vect.transform(x_test)
```

```python
# Classifier - Algorithm - SVM
# fit the training dataset on the classifier
SVM = svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='auto')
SVM.fit(Train_X_Tfidf,y_train)# predict the labels on validation dataset
predictions_SVM = SVM.predict(Test_X_Tfidf)# Use accuracy_score function to get the accuracy
print("SVM Accuracy Score -> ",accuracy_score(predictions_SVM, y_test)*100)
```

```python
Naive = naive_bayes.MultinomialNB()
Naive.fit(Train_X_Tfidf,y_train)# predict the labels on validation dataset
predictions_NB = Naive.predict(Test_X_Tfidf)# Use accuracy_score function to get the accuracy
print("Naive Bayes Accuracy Score -> ",accuracy_score(predictions_NB, y_test)*100)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_Final['News'] ,df_Final['Close'].values,test_size=0.2)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tok=Tokenizer(3000)

tok.fit_on_texts(x_train)
x_train_se=tok.texts_to_sequences(x_train)

x_test_se=tok.texts_to_sequences(x_test)
```

```python
from keras.preprocessing.sequence import pad_sequences
x_train=pad_sequences(x_train_se,maxlen=150)
x_test=pad_sequences(x_test_se,maxlen=150)
```

```python
from tensorflow.keras import Model
from tensorflow.keras.layers import Input,LSTM,Dense,Activation,Dropout
from keras.layers.embeddings import Embedding 
```

```python
max_len=150
def rnn():
  inputs =Input(name='input',shape=[max_len])
  layer=Embedding(3000,50,input_length=max_len)(inputs)
  layer=LSTM(64)(layer)
  layer=Dense(256,name='FC1')(layer)
  layer=Activation('relu')(layer)
  layer=Dropout(0.5 )(layer)
  layer=Dense(1,name='out_layer')(layer)
  layer=Activation('relu')(layer)
  model=Model(inputs=inputs,outputs=layer)
  return model
```

```python
model=rnn()
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=10,batch_size=64,validation_split=0.2)
```
